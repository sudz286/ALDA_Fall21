```python
# Function to normalize Training and Testing data sets
from sklearn import preprocessing
import pandas as pd

def normalize_df(train_df,test_df):
    new_train_df = train_df.drop('Class', axis = 1)
    new_test_df = test_df.drop('Class', axis = 1)
    for x in new_train_df.columns:
        new_train_df[x] = (new_train_df[x] - new_train_df[x].min()) / (new_train_df[x].max() - new_train_df[x].min())
    for x in new_test_df.columns:
        new_test_df[x] = (new_test_df[x] - new_train_df[x].min()) / (new_train_df[x].max() - new_train_df[x].min())
    return new_train_df, new_test_df

norm_train,norm_test = normalize_df(train,test)
y_train = train['Class']
y_test = test['Class']

norm_train.head()
```

```python
def covar(x):
    #Calculating covariance matrix of the training dataset
    return x.cov()
    # DDOF = 0 gives us the population covariance
    
cov_train = covar(norm_train)
cov_test = covar(norm_test)
print('Dimensions of the Training covariance matrix is', cov_train.shape)
cov_train[cov_train.columns[:5]].head(5)
```

```python
from numpy.linalg import eig

def det_eig_df(x,norm):
    cols = norm.columns
    eig_va, eig_ve = eig(x)
    eig_m = pd.DataFrame(cols,columns = ['attributes'])
    eig_m['eigen values'] = eig_va
    eig_m['eigen vectors'] = [x for x in eig_ve]
#     eig_m.sort_values(by = 'eigen values', ascending = False)
    return eig_m,eig_ve

eigen_list_train,eigen_list_train_vectors = det_eig_df(cov_train,norm_train)
eigen_list_test, eigen_list_test_vectors = det_eig_df(cov_test, norm_test)

eigen_list_train
```

```python
import matplotlib.pyplot as plt
def plot_eigen(eigen_matrix):
    eig_30 = eigen_matrix['eigen values']
    fig = plt.figure()
    fig.set_figwidth(20)
    fig.set_figheight(10)
    plt.xlabel('Number of Eigenvalues')
    plt.ylabel('Eigenvalue')
    plt.plot(eig_30)
    plt.show()

plot_eigen(eigen_list_train)
```

```python
# Determing the percentage of variance accounted for by each of the first 30 components
def eig_variance(e):
    eigen_sum = 0
    eigen_sum = e['eigen values'].sum()
    for i in range(30):
        print((e['eigen values'][i]/eigen_sum)*100)
    # We have chosen the first three eigenvectors for PCA since it covers nearly 83% of the original data
    
eig_variance(eigen_list_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

def fea_vec_calc(X,eig,p):
    feature_vector = np.array([ eig['eigen vectors'][i] for i in range(p)])
    print(X.shape)
    print(feature_vector.T.shape)
    result = np.dot(X,feature_vector.T)
    return result

def knn_p(train_set,test_set,e_train,e_test,):
    accuracy_list = []
    p_values = [2,4,8,10,20,25,30]
    for p in p_values:
        train_components = fea_vec_calc(train_set,e_train,p)
        test_components = fea_vec_calc(test_set,e_test,p)
        neigh = KNeighborsClassifier(n_neighbors=5)
        neigh.fit(train_components, train['Class'])
        predicted = neigh.predict(test_components)
        accuracy = accuracy_score(test['Class'],predicted)
        print('Accuracy when p = ',p,'->',accuracy)
        accuracy_list.append(accuracy)

    plt.xlabel('Number of Principal Components')
    plt.ylabel('Accuracy of testing dataset')
    plt.plot(p_values,accuracy_list)

knn_p(norm_train,norm_test,eigen_list_train,eigen_list_test)
```
